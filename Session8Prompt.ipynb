```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 1, 100)

#Sin function
def sinX(x):
    return np.sin(x)

#Cos function
def cosX(x):
    return np.cos(x)
```

```python
plt.figure(figsize=(10, 5))

#Left panel: sin(x) vs x
plt.subplot(1, 2, 1)
plt.plot(x, sinX(x), label='sin(x)', color='blue')
plt.title('sin(x) vs x')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.grid(True)
plt.legend()

#Right panel: cos(x) vs x
plt.subplot(1, 2, 2)
plt.plot(x, cosX(x), label='cos(x)', color='red')
plt.title('cos(x) vs x')
plt.xlabel('x')
plt.ylabel('cos(x)')
plt.grid(True)
plt.legend()

#Save the figure as a PDF
plt.savefig('session8.pdf', bbox_inches='tight', facecolor='white')
```
