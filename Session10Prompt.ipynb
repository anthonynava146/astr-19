```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Log series random
nums = np.random.normal(.6, 1000)

#Create a histogram
plt.figure(figsize=(10, 6))
plt.hist(nums, bins=100, color='blue', edgecolor='black')
plt.xlabel('Value')
plt.ylabel('# of Values')
plt.title('Histogram')

#Saves the histogram
plt.savefig('session9.pdf', bbox_inches='tight', facecolor='white')
```
