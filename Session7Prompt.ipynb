```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#x array between [0,1] 
x = np.linspace(0, 1, 100)

#exponent function
def exp(x):
    return np.exp(x)

#sets y to the function
y = exp(x)

#draws y and lables axes
plt.plot(x, y)
plt.xlabel('Time [milliseconds]')
plt.ylabel('Awesomeness')
plt.title('Exponential Growth')

#saves figure as a pdf
plt.savefig('Session7Prompt.pdf')

```
